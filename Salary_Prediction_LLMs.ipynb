```python
# dropping all the null values from the dataset
df1 = df.dropna()
df1.isnull().sum()
```

```python
# Compare transformed source data with target data
validation = df[['Age', 'Job Title']].merge(df1, on=['Age', 'Job Title'], how='outer', indicator=True)

# Identify mismatches
mismatches = validation[validation['_merge'] != 'both']

if mismatches.empty:
    print("ETL transformation is accurate.")
else:
    print("Discrepancies found in ETL transformation:")
    print(mismatches)
```

```python
# Accuracy: Validate transaction amounts against a source system (example assumes a validation function exists)
def validate_amounts(df):
    # Placeholder for validation logic
    return df['Salary'].apply(lambda x: True if x >= 0 else False)

df['amount_valid'] = validate_amounts(df)

# Completeness: Check for missing values
completeness = df.isnull().sum()

# Consistency: Check for unique transaction IDs
consistency = df['Job Title'].is_unique

# Timeliness: Measure ETL process duration (example assumes start and end times are tracked)
import time
start_time = time.time()
# ETL process...
end_time = time.time()
timeliness = end_time - start_time

# Uniqueness: Check for duplicates
duplicates = df.duplicated().sum()

# Log metrics
metrics = {
    'accuracy': df['amount_valid'].mean(),
    'completeness': completeness,
    'consistency': consistency,
    'timeliness': timeliness,
    'uniqueness': duplicates
}

print(metrics)
```

```python
# Define the chunk size
chunk_size = 1000  # Number of rows per chunk

# Initialize an empty list to collect processed data
processed_data = []

# Iterate over the CSV file in chunks
for chunk in pd.read_csv('Salary Data.csv', chunksize=chunk_size):
    # Perform some processing on each chunk
    # For example, let's say we want to calculate the mean of the 'value' column
    chunk_mean = chunk['Salary'].mean()
    print(f"Chunk mean: {chunk_mean}")

    # Append the processed result to the list
    processed_data.append(chunk_mean)

# After processing all chunks, you can further process the collected data
overall_mean = sum(processed_data) / len(processed_data)
print(f"Overall mean: {overall_mean}")
```

```python
# Check for missing values
print("Missing values before handling:")
print(df.isnull().sum())

# Fill missing values with a default value
df_filled_default = df.fillna({
    'transaction_id': -1,
    'customer_id': -1,
    'transaction_date': '1970-01-01',
    'amount': 0.0
})

print("Missing values after filling with default values:")
print(df_filled_default.isnull().sum())

# Save the DataFrame with default values to a new CSV file
df_filled_default.to_csv('Salary Data.csv', index=False)

# Fill missing values with the mean of the column (only for numeric columns)
df_filled_mean = df.copy()
df_filled_mean['Salary'].fillna(df['Salary'].mean(), inplace=True)

print("Missing values after filling 'amount' with mean:")
print(df_filled_mean.isnull().sum())

# Save the DataFrame with mean-filled values to a new CSV file
df_filled_mean.to_csv('filled_mean_financial_transactions.csv', index=False)

print("Missing values handled and data saved to respective files.")
```

```python
# Converting the Categorial data into numbers
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df1['Education Level'] = label_encoder.fit_transform(df1['Education Level'])
df1.head()
```

```python
# Splitting the data into x and y for training and testing purpose
x=df1[['Education Level', 'Years of Experience']]
x.shape
```

```python
x.head(5)
```

```python
y=df1[['Salary']]
y.head(5)
```

```python
# Splitting the above data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 9598)
```

```python
# to know about the shape of traing and testing data

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# As it is the Regrssion Problem lets try all the Regression Algorith to get the best accuracy
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# To train and fit the model
model.fit(x_train,y_train)
```

```python
# to predict the data
y_pred = model.predict(x_test)
y_pred
```

```python
# to know what the orignal prediction was for the testing data
y_test
```

```python
# to Calculate the Accuracy of the Model
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error,mean_squared_error
```

```python
# to print the mean absolute error
mean_absolute_error(y_test,y_pred)
```

```python
# to print the mean absolute percentage
per_e = mean_absolute_percentage_error(y_test,y_pred)
per_e
```

```python
# to print the accuracy in percentage
acc1 = (1-per_e)*100
acc1
```

```python
# Trying other Algorhithms to check which gives the best accuracy

# KNeighbors Regressor

from sklearn.neighbors import KNeighborsRegressor
model2 = KNeighborsRegressor()
model2.fit(x_train,y_train)
y_pred2 = model2.predict(x_test)
error2 = mean_absolute_percentage_error(y_test,y_pred2)
acc2 = (1-error2)*100
acc2
```

```python
# DecisionTree Regressor

from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor()
model3.fit(x_train,y_train)
y_pred3 = model3.predict(x_test)
error3 = mean_absolute_percentage_error(y_test,y_pred3)
acc3 = (1-error3)*100
acc3
```

```python
# RandomForest Regressor

from sklearn.ensemble import RandomForestRegressor
model4 = RandomForestRegressor()
model4.fit(x_train,y_train)
y_pred4 = model4.predict(x_test)
error4 = mean_absolute_percentage_error(y_test,y_pred4)
acc4 = (1-error4)*100
acc4
```

```python
print("The Acuuracy of Linear Regressor is ",acc1)
print("The Acuuracy of KNeighbors Regressor is ",acc2)
print("The Acuuracy of DecisionTree Regressor is ",acc3)
print("The Acuuracy of RandomForest Regressor is ",acc4)
```
